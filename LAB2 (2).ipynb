```python
# no missing values to fill
thyroid.isna().sum()
```

```python
missing_values = thyroid.isnull().sum()
print(missing_values)
```

```python
numeric_columns = thyroid.select_dtypes(include=['float64', 'int64']).columns
print(thyroid[numeric_columns].describe())
```

```python
median=thyroid['Record ID'].median()
mean_age=thyroid['age'].mean()
mean_age
```

```python
thyroid['age'].fillna(mean_age,inplace=True)
thyroid['Record ID'].fillna(median,inplace=True)
```

```python

# Impute other numeric columns with median
for col in numeric_columns:
    if col == 'Record ID':
       print('checked')
       continue
    thyroid[col] = thyroid[col].fillna(thyroid[col].median())

# Impute categorical columns with mode
categorical_columns = thyroid.select_dtypes(include=['object']).columns
for col in categorical_columns:
    thyroid[col] = thyroid[col].fillna(thyroid[col].mode()[0])
```
